```python

plt.figure(figsize = (12, 6)) # Ancho, Alto del grafico

marcas_clase = [0,1,2,3,4,5,6,7] # Datos del eje X
frecuencias = [405, 411, 445, 1016, 1091, 1370, 3736, 7912] # Datos del eje Y
marcas_texto = ["Sedán", "Pointer", "Estaquitas", "Bora", "Jetta 4a", "Sentra", "Pick up", "Tsuru"]

# El histograma tiene: 
# - Barras al 100% (width= 1)
# - Color de contorno negro (edgecolor= "k")
# - Seleccionar colo hexadecimal: https://htmlcolorcodes.com/es/
plt.barh(marcas_clase,frecuencias, 
        height= 0.7, edgecolor= "k", 
        color =["#3DE7E1","#F532DD","#63FC44","#44FCE3","#4495FC","#9844FC","#F444FC","#FC44AE"])

plt.yticks(marcas_clase, marcas_texto, fontsize = 15)
plt.xlabel("Frecuencia absoluta", fontsize=25) # Etiqueta del eje X
plt.ylabel("Marcas de clase", fontsize=25) # Etiqueta del eje Y
plt.title("Diagrama de barras", fontsize=35) # Titulo
plt.grid() # Activar cuadricula
plt.show() # Mostrar el grafico
```
